## ML Concepts Covered
- Binary classification output: flag (1) vs no-flag (0)
- Confusion matrix: TP, FP, TN, FN
- Precision = TP / (TP + FP)
- Recall = TP / (TP + FN)
- F1 = 2 x (Precision x Recall) / (Precision + Recall)
