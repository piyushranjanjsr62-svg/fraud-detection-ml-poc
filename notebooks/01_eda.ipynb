## 1. Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
plt.rcParams['figure.figsize'] = (12, 5)
sns.set_style('whitegrid')

print('All libraries loaded successfully.')
```

## 2. Load the Dataset

```python
customers    = pd.read_csv('../data/raw/customers.csv')
accounts     = pd.read_csv('../data/raw/accounts.csv')
transactions = pd.read_csv('../data/raw/transactions.csv')

print('customers.csv    :', customers.shape)
print('accounts.csv     :', accounts.shape)
print('transactions.csv :', transactions.shape)
```

## 3. Verify Approved Dataset Parameters
These values are locked in V5.0 Section 5.2. If any check fails, stop and investigate.

```python
print('=== DATASET PARAMETER VERIFICATION ===')
print()

# Transaction count
txn_count = len(transactions)
status = 'PASS' if txn_count == 50800 else 'FAIL'
print(f'Total transactions : {txn_count:,}  (expected 50,800)  [{status}]')

# Customer count
cust_count = len(customers)
status = 'PASS' if cust_count == 500 else 'FAIL'
print(f'Total customers    : {cust_count:,}  (expected 500)  [{status}]')

# Account count
acc_count = len(accounts)
status = 'PASS' if acc_count == 780 else 'FAIL'
print(f'Total accounts     : {acc_count:,}  (expected 780)  [{status}]')

# Fraud count and rate
fraud_count = transactions['is_fraud'].sum()
fraud_rate  = fraud_count / txn_count
status_count = 'PASS' if fraud_count == 254 else 'CHECK'
status_rate  = 'PASS' if 0.004 <= fraud_rate <= 0.006 else 'FAIL'
print(f'Fraud cases        : {fraud_count}  (expected 254)  [{status_count}]')
print(f'Fraud rate         : {fraud_rate:.4f} = {fraud_rate*100:.2f}%  (expected ~0.5%)  [{status_rate}]')

print()
print('=== VERIFICATION COMPLETE ===')
```

## 4. Temporal Range Check
Critical for defining the 70/20/10 temporal split boundaries.

```python
transactions['timestamp'] = pd.to_datetime(transactions['timestamp'])

print('=== TEMPORAL RANGE ===')
print(f"Earliest transaction : {transactions['timestamp'].min()}")
print(f"Latest transaction   : {transactions['timestamp'].max()}")
print(f"Date span            : {(transactions['timestamp'].max() - transactions['timestamp'].min()).days} days")

# Sort by time to prepare for split
transactions = transactions.sort_values('timestamp').reset_index(drop=True)
print()
print('Transactions sorted by timestamp.')

# Calculate split boundaries
train_end = int(len(transactions) * 0.70)
val_end   = int(len(transactions) * 0.90)

print(f"\n=== TEMPORAL SPLIT BOUNDARIES ===")
print(f"Train (70%) : rows 0 – {train_end:,}  | cutoff date: {transactions.iloc[train_end]['timestamp']}")
print(f"Val   (20%) : rows {train_end:,} – {val_end:,}  | cutoff date: {transactions.iloc[val_end]['timestamp']}")
print(f"Test  (10%) : rows {val_end:,} – {len(transactions):,}")
```

## 6. Missing Values Check

```python
print('=== MISSING VALUES — TRANSACTIONS ===')
missing = transactions.isnull().sum()
missing_pct = (missing / len(transactions) * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
print(missing_df[missing_df['Missing Count'] > 0].to_string())

if missing_df['Missing Count'].sum() == 0:
    print('No missing values found in transactions.csv')

print()
print('=== MISSING VALUES — CUSTOMERS ===')
cust_missing = customers.isnull().sum()
print(cust_missing[cust_missing > 0].to_string() if cust_missing.sum() > 0 else 'No missing values.')
```

## 7. Duplicate Check

```python
dup_count = transactions.duplicated(subset=['transaction_id']).sum()
print(f'Duplicate transaction_ids : {dup_count}')

if dup_count > 0:
    print('WARNING: Duplicates found — investigate before training')
else:
    print('PASS — No duplicate transaction IDs')
```

## 12. Submission Hour Analysis (Off-Hours Fraud Signal)

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Hour of day: fraud vs legitimate
legit_hours = transactions[transactions['is_fraud']==0]['hour_of_day']
fraud_hours = transactions[transactions['is_fraud']==1]['hour_of_day']

axes[0].hist(legit_hours, bins=24, alpha=0.6, label='Legitimate', color='steelblue', density=True)
axes[0].hist(fraud_hours, bins=24, alpha=0.7, label='Fraud',      color='crimson',   density=True)
axes[0].set_title('Hour of Day: Fraud vs Legitimate')
axes[0].set_xlabel('Hour of Day (0-23)')
axes[0].set_ylabel('Density')
axes[0].legend()

# Day of week: fraud count
fraud_by_dow = transactions[transactions['is_fraud']==1]['day_of_week'].value_counts().sort_index()
dow_labels = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
axes[1].bar([dow_labels[i] for i in fraud_by_dow.index], fraud_by_dow.values, color='crimson')
axes[1].set_title('Fraud Cases by Day of Week')
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Fraud Count')

plt.tight_layout()
plt.show()
```

## 13. ⚠️ LABEL LEAKAGE GATE
**Rule from V5.0 Section 5.5:** Train a Logistic Regression using only 3 simple features.  
- If AUC > 0.90 → **STOP. Do not proceed. The dataset has a leakage problem.**  
- If AUC < 0.90 → **PASS. Safe to proceed to feature engineering.**

```python
print('=== LABEL LEAKAGE GATE ===')
print('Running 3-feature Logistic Regression smoke test...')
print()

# Use only 3 simple features that should NOT perfectly predict fraud
leakage_features = ['amount_usd', 'hour_of_day', 'day_of_week']

# Check all features exist
available = [f for f in leakage_features if f in transactions.columns]
print(f'Features used: {available}')

df_leak = transactions[available + ['is_fraud']].dropna()

X = df_leak[available].values
y = df_leak['is_fraud'].values

lr = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
lr.fit(X, y)
probs = lr.predict_proba(X)[:, 1]
auc = roc_auc_score(y, probs)

print(f'\n3-Feature LR ROC-AUC : {auc:.4f}')
print()

if auc > 0.90:
    print('FAIL — AUC > 0.90')
    print('STOP: Likely label leakage in the dataset. Do not proceed to modelling.')
    print('Action: Review dataset generator for features that encode is_fraud directly.')
else:
    print('PASS — AUC < 0.90')
    print('No obvious label leakage detected.')
    print('Safe to proceed to Feature Engineering (Week 2).')
```
