```python
print("Mean:  ", df['amount_usd'].mean().round(2))
print("Median:", df['amount_usd'].median().round(2))
print("Skewness:", df['amount_usd'].skew().round(3))
```

```python
Q1 = df['amount_usd'].quantile(0.25)
Q3 = df['amount_usd'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['amount_usd'] < lower) | (df['amount_usd'] > upper)]
print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
print(f"Lower fence: {lower:.2f}")
print(f"Upper fence: {upper:.2f}")
print(f"Outlier count: {len(outliers)}")
```

```python
print("Total transactions:", len(df))
print("Fraud cases:", df['is_fraud'].sum())
print("Fraud rate:", (df['is_fraud'].mean() * 100).round(3), "%")
```

```python
df.groupby('is_fraud')['amount_usd'].mean().round(2)
```

```python
fraud_by_biz = df.groupby('business_type')['is_fraud'].agg(
    total='count',
    fraud_cases='sum',
    fraud_rate='mean'
).round(4)

fraud_by_biz['fraud_rate_pct'] = (fraud_by_biz['fraud_rate'] * 100).round(2)
print(fraud_by_biz.sort_values('fraud_rate_pct', ascending=False))
```

```python
fraud_by_corridor = df.groupby('beneficiary_country')['is_fraud'].agg(
    total='count',
    fraud_cases='sum',
    fraud_rate='mean'
)
fraud_by_corridor['fraud_rate_pct'] = (fraud_by_corridor['fraud_rate'] * 100).round(2)
print(fraud_by_corridor.sort_values('fraud_rate_pct', ascending=False).head(10))
```

```python
fraud_by_hour = df.groupby('hour_of_day')['is_fraud'].mean() * 100
print(fraud_by_hour.round(3))
```

```python
bins = [0, 10000, 25000, 50000, 100000, 200000]
labels = ['0-10K', '10K-25K', '25K-50K', '50K-100K', '100K-200K']

df['amount_band'] = pd.cut(df['amount_usd'], bins=bins, labels=labels)

fraud_by_band = df.groupby('amount_band', observed=True)['is_fraud'].agg(
    total='count',
    fraud_cases='sum',
    fraud_rate='mean'
)
fraud_by_band['fraud_rate_pct'] = (fraud_by_band['fraud_rate'] * 100).round(2)
print(fraud_by_band)
```
